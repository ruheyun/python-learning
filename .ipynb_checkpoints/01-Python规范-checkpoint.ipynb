Python 规范

注释规则：

    使用 # 表示单行注释，建议放到需要注释代码上单独一行，或者放到代码同行后面，注释和语句空两格；
    
    使用三个单引号或三个双引号表示多行注释。

代码缩进：

    Python 采用代码缩进和冒号来区分代码之间的层次；
    
    建议使用 4 个空格作为一个缩进量。

标点符号：

    所有标点符号都需要在英文输入法下输入。

```python
print('Hello world!')  # 这是一个单行注释，注释内容不会被编译和输出
```

```python
a = 1
b = 1
if a == b:  # 这里冒号需要在英文输入法下输入
    print('演示')  # 这里用 4 个空格区分代码层次
```
